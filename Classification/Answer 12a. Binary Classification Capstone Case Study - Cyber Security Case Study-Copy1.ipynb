```python
actual_train=train.Attack_Binary
pred_prob_train=model.predict(train)
```

```python
# determining the best cut_off and prediction

roc_df =pd.DataFrame()
for cut_off in np.linspace(0,1,50):
    TP = np.where(((actual_train==1) & (np.where(pred_prob_train>cut_off,1,0)==1)),1,0)
    TN = np.where(((actual_train==0) & (np.where(pred_prob_train>cut_off,1,0)==0)),1,0)
    FP = np.where(((actual_train==0) & (np.where(pred_prob_train>cut_off,1,0)==1)),1,0)
    FN = np.where(((actual_train==1) & (np.where(pred_prob_train>cut_off,1,0)==0)),1,0)

    sensitivity = TP.sum()/(TP.sum()+ FN.sum())
    spcificity = TN.sum()/(TN.sum()+ FP.sum())
    total = sensitivity+ spcificity
    accuracy = (TP.sum() + TN.sum())/(TP.sum() + TN.sum()+ FP.sum() + FN.sum())
    
    temp = pd.DataFrame([cut_off, sensitivity, spcificity, total, accuracy]).T
    roc_df = pd.concat([roc_df, temp], axis=0)
    
    print('Cut-off:', cut_off)
    print('######################################')
    print('sensitivity:', sensitivity)
    print('spcificity:', spcificity)
    print('total:', total)
```

```python
roc_df.columns=['cut_off', 'sensitivity', 'spcificity', 'total', 'accuracy']
roc_df.sort_values(by='total',ascending=False)
```

```python
cut_off_value=roc_df.sort_values(by='total',ascending=False).head(1)['cut_off'][0]
```

```python
cut_off_value
```

```python
train['pred_prob']=model.predict(train)
test['pred_prob']=model.predict(test)
```

```python
train['pred_val']=np.where(train.pred_prob>cut_off_value,1,0)
test['pred_val']=np.where(test.pred_prob>cut_off_value,1,0)
```
